
# Homework: Python for loops and conditional statements 

This homework will test your understanding of Python for loops, basic operations, and conditional statements. Each problem is designed to challenge you and encourage exploration of Python's capabilities.

---

## Problem 1: Counting Nucleotides
Write a program that takes a DNA sequence (e.g., "ATGCGCAT") as input and counts the number of occurrences of each nucleotide (A, T, G, C).

### Example Input:
```
ATGCGCAT
```

### Example Output:
```
A: 2
T: 2
G: 2
C: 2
```

---

## Problem 2: Transcribing DNA to RNA
Write a program that takes a DNA sequence as input and outputs the corresponding RNA sequence by replacing every occurrence of "T" with "U."

### Example Input:
```
ATGCGCAT
```

### Example Output:
```
AUGCGCAU
```

---

## Problem 3: Complementary Strand
Write a program to generate the complementary DNA strand for a given DNA sequence. Use the following rules:
- A pairs with T
- T pairs with A
- C pairs with G
- G pairs with C

### Example Input:
```
ATGCGCAT
```

### Example Output:
```
TACGCGTA
```

---

## Problem 4: GC Content Calculator
Write a program that calculates the GC content (percentage of G and C) in a given DNA sequence. The result should be rounded to two decimal places.

### Example Input:
```
ATGCGCAT
```

### Example Output:
```
GC Content: 50.00%
```

---

## Problem 5: Reverse a DNA Sequence
Write a program that takes a DNA sequence as input and outputs the sequence in reverse order.

### Example Input:
```
ATGCGCAT
```

### Example Output:
```
TACGCGTA
```
