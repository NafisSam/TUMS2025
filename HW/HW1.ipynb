
# Homework: Python Basics and Strings

This homework will test your understanding of Python variables, basic operations, and string manipulations. Each problem is designed to challenge you and encourage exploration of Python's capabilities.

---

## Problem 1: String Manipulations and Formatting

Write a Python program that takes a user's first and last name as input and does the following:
1. Converts the first name to all uppercase letters.
2. Converts the last name to all lowercase letters.
3. Combines the first and last name into a single string with the format: `"Last Name, First Name"`.
4. Prints the formatted string.

Example:
```plaintext
Input:
First Name: John
Last Name: DOE
Output:
doe, JOHN
```

---

## Problem 2: Arithmetic Operations and Input

Write a program that:
1. Asks the user for two numbers (you can assume valid inputs).
2. Computes and prints:
   - The sum of the numbers.
   - The product of the numbers.
   - The division result (as a float) of the first number divided by the second.
3. Format the results to show up to 2 decimal places.

Example:
```plaintext
Input:
Enter first number: 10
Enter second number: 3
Output:
Sum: 13
Product: 30
Division: 3.33
```

---

## Problem 3: Counting and String Analysis

Write a program that:
1. Takes a string input from the user.
2. Counts and prints the number of vowels in the string.
3. Converts the string to uppercase and prints the result.
4. Replaces all spaces in the string with a dash (`-`) and prints the modified string.

Example:
```plaintext
Input:
Hello World
Output:
Vowel Count: 3
Uppercase: HELLO WORLD
Modified String: HELLO-WORLD
```

---

## Problem 4: Palindrome Checker

A palindrome is a string that reads the same backward as forward. Write a program that:
1. Asks the user for a string.
2. Checks if the string is a palindrome (case-insensitive).
3. Prints `"Palindrome"` if true, otherwise `"Not a palindrome"`.

Example:
```plaintext
Input:
Racecar
Output:
Palindrome
```

---

## Problem 5: Advanced String Formatting

Write a program that:
1. Prompts the user to enter a sentence.
2. Formats the sentence to:
   - Capitalize the first letter of every word.
   - Center-align the sentence within a width of 50 characters, filling the space with `*`.
3. Prints the formatted sentence.

Example:
```plaintext
Input:
python is awesome
Output:
***********Python Is Awesome************
```
